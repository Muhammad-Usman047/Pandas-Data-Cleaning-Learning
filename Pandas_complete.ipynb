```python
df.loc[df["Monthly Salary (INR)"].notna(), "Status"] = "EMPLOYED"
df.loc[df["Monthly Salary (INR)"] == 0.0, "Status"] = "UNEMPLOYED"
```

```python
df["Education"].value_counts(dropna=False)
```

```python
df["Education"] = df["Education"].replace({
    "Bachelors": "Bachelor",
    "Bachelor's": "Bachelor",
    "Masters": "Master",
    "Master": "Master",
    "PhD": "PhD",
    "Ph.D": "PhD",
    "High School": "High School",
    "high school": "High School",
    "None": None
})
```

```python
df["Education"].value_counts(dropna = False)
```

```python
df["Education"] = df["Education"].replace("NaN", np.nan)
```

```python
df["Education"].value_counts(dropna = False)
```

```python
df["Education"] = df["Education"].fillna(df["Education"].mode()[0])
df
```

```python
df.isnull().sum()
```

```python
df["Industry"].value_counts(dropna = False)
```

```python
df["Industry"] = df["Industry"].replace(" Fintech", "Fintech")
```

```python
df["Industry"].value_counts(dropna = False)
```

```python
df["Industry"] = df["Industry"].fillna(df["Industry"].mode()[0])
df["Location"] = df["Location"].fillna(df["Location"].mode()[0])
df["AI Risk"] = df["AI Risk"].fillna(df["AI Risk"].mode()[0])
df
```

```python
df["AI Risk"] = df["AI Risk"].replace({
    "low" : "Low",
    "moderate" : "Moderate",
    "Medium" : "Moderate",
    "HIGH": "High"
})
df["AI Risk"].value_counts(dropna = False)
```

```python
df["Age Group"] = df["Age Group"].replace({
    "18 to 24":"18-24",       
    "25_34":"25-34",
    "35 - 44":"35-44" 
})
df["Age Group"].value_counts(dropna = False)
```
